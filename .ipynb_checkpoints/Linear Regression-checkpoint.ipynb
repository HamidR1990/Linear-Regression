```python
#################### Linear Regression ####################
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer


#################### Hyperparameters
epoch = 200
etta = 0.01
train_rate = 0.7
val_rate = 0.1
test_rate = 0.2


#################### Loading Dataset and Preprocessing
df = pd.read_csv("data/usa-housing.csv")

col_dic = {"Avg. Area Income": "Avg. Income", "Avg. Area House Age": "Avg. Age", "Avg. Area Number of Rooms": "Avg. Rooms",
          "Avg. Area Number of Bedrooms": "Avg. Bedrooms", "Area Population": "Population"}

df.rename(columns=col_dic, inplace=True)
df = df.drop(['Address'], axis = 1)
data = np.array(df)
M,N = np.shape(data)



def standardisation(data):
    out = (data - np.mean(data, axis=0))/ np.std(data, axis=0)
    return out


def normalisation(data):
    m = np.min(data, axis=0)
    M = np.max(data, axis=0)
    out = (data - m) / (M - m)
    return out


data = standardisation(data)
data = np.hstack((np.ones((M,1)), data))

```

```python
#################### Train-Test-Validation Split

M_train = int(np.ceil(train_rate * M))
M_val = int(np.ceil(val_rate * M))
M_test = M - (M_train + M_val)


idx = np.random.permutation(M)
train_idx = idx[:M_train]
val_idx = idx[M_train: M_val+M_train]
test_idx = idx[M_val+M_train:]

train_data = data[train_idx, :]
val_data = data[val_idx, :]
test_data = data[test_idx, :]

X_train = train_data[:, :-1]
Y_train = train_data[:, -1]
X_val = val_data[:, :-1]
Y_val = val_data[:, -1]
X_test = test_data[:, :-1]
Y_test = test_data[:, -1]
```

```python
#################### Linear Regression Model

def MSE(Y, Y_hat, n):
    error = (1/n) * (Y - Y_hat)**2
    out = np.sum(error) 
    return out


def RMSE(Y, Y_hat, n):
    error = (1/n) * (Y - Y_hat)**2
    out = np.sum(error) 
    return np.sqrt(out)


def MAE(Y, Y_hat, n):
    error = (1/n) * np.abs(Y - Y_hat)
    out = np.sum(error)
    return out


def R2(Y, Y_hat, n):
    s1 = np.sum((Y - Y_hat)**2)
    s2 = np.sum((Y - np.mean(Y_hat))**2)
    out = 1 - (s1/s2)
    return out


def grad(X, Y, Y_hat, n):
    grad = (2/n) * (np.transpose(X) @ (Y_hat - Y))
    return grad




W = np.random.rand(N)
L_train = np.array([])
L_val = np.array([])


for i in range(epoch):
    Y_hat_train = X_train @ W
    Y_hat_val = X_val @ W
    
    c_train = MSE(Y_train, Y_hat_train, M_train)
    L_train = np.append(L_train, c_train)
    c_val = MSE(Y_val, Y_hat_val, M_val)
    L_val = np.append(L_val, c_val)
    
    W = W - (etta * grad(X_train, Y_train, Y_hat_train, M_train)) 


```

```python
#################### Learning Curve

fig, ax = plt.subplots()

plt.plot(range(1, epoch+1), L_train, label='Train')
plt.plot(range(1, epoch+1), L_val, label='Validation')
plt.title("Learning Curve")
plt.xlabel("Error")
plt.ylabel("Epoch")
plt.grid()
plt.legend(loc="lower left")

txt = '\n'.join((
    r'MSE Train=%.2f' % (MSE_train, ),
    r'MSE Validation=%.2f' % (MSE_val, ),
    r'MSE Test=%.2f' % (MSE_test, ),
    r'R2 Test=%.2f' % (r2_test, )))

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

plt.text(0.97, 0.96, txt, transform=ax.transAxes, fontsize=10, verticalalignment='top', horizontalalignment='right', bbox=props)



plt.show()

print("The MSE of Train Is: ", MSE_train)
print("The MSE of Validation Is: ", MSE_val)
print("The MSE of Test Is: ", MSE_test)
print("The R2-Score of Test is: ", r2_test)
```
